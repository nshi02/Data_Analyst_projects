#Проверка гипотез

Н0: Пользователи, отписавшиеся от коммуникаций (email и звонков), бронируют отели с тем же уровнем использования промокодов и кэшбэка, как и пользователи, не отписавшиеся.\
Н1:Пользователи, отписавшиеся от коммуникаций, отличаются по уровню использования промокодов и кэшбэка при бронировании отелей по сравнению с теми, кто не отписался.

```python
df['is_organic'] = ((df['promo_code_discount_amt'].fillna(0) == 0) &
                    (df['loyalty_accrual_rub_amt'].fillna(0) == 0)).astype(int)


df['opted_out'] = ((df['suppress_email_flg'] == 1) & (df['suppress_call_flg'] == 1)).astype(int)


organic_share = df.groupby('opted_out')['is_organic'].mean()

print("0 — не отписались, 1 — отписались")
print(organic_share)
```

```python
from scipy.stats import mannwhitneyu

group_0 = df[df['opted_out'] == 0]['is_organic']
group_1 = df[df['opted_out'] == 1]['is_organic']

stat, p_value = mannwhitneyu(group_0, group_1, alternative='two-sided')

print(f"p-value: {p_value:.5f}")
```

p-value>0.05 -> нет причин отвергнуть нулевую гипотезу. Т.о. отказ от коммуникаций не влияет на использование промокодов и кэшбека

H0:Пользователи, участвующие в программе лояльности, совершают заказы с такой же частотой, как и те, кто не участвует. \
Н1:Пользователи, участвующие в программе лояльности, совершают заказы чаще, чем те, кто не участвует.\
Проверка данной гипотезы показывает необходимость вовлечения клиентов в программу лояльности.

```python
from scipy import stats
loyalty_users = df[df['loyalty_program_type_nm'].notna()]
no_loyalty_users = df[df['loyalty_program_type_nm'].isna()]


orders_loyalty = loyalty_users.groupby('client_rk')['order_rk'].nunique()
orders_no_loyalty = no_loyalty_users.groupby('client_rk')['order_rk'].nunique()


t_stat, p_value = stats.ttest_ind(orders_loyalty, orders_no_loyalty)

print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')

if p_value < 0.05:
    print("Отклоняем нулевую гипотезу: Участвующие в программе лояльности заказывают чаще.")
else:
    print("Не можем отклонить нулевую гипотезу: Нет статистически значимой разницы.")
```

H0: Нет связи между уровнем дохода клиента и частотой заказов.

Н1: Есть связь между уровнем дохода клиента и частотой заказов.

```python
from scipy import stats
df['monthly_income_amt'] = pd.to_numeric(df['monthly_income_amt'], errors='coerce')
income_order_counts = df.groupby('monthly_income_amt')['order_rk'].nunique()
low_income = income_order_counts[income_order_counts.index < 50000]  # условный порог
high_income = income_order_counts[income_order_counts.index >= 50000]

t_stat_1, p_value_1 = stats.ttest_ind(low_income, high_income)

print(f"Гипотеза 1: t-statistic: {t_stat_1}, p-value: {p_value_1}")
if p_value_1 < 0.05:
    print("Отклоняем нулевую гипотезу: Есть связь между уровнем дохода и частотой заказов.")
else:
    print("Не отклоняем нулевую гипотезу: Нет связи между уровнем дохода и частотой заказов.")

```

H0: Местоположение клиента не влияет на частоту или вероятность совершения покупки. \
H1: Местоположение клиента влияет на частоту или вероятность совершения покупки. \
Данная гипотеза показывает актуальность предоставления кэшбеков и скидок для жителей в зависимости от регионов, с целью повышения вовлеченности клиентов из регионов с низкой вовлеченностью. Можно также применить персонализированные рекоммендации.

```python
from scipy import stats
region_order_counts = df.groupby('lvn_state_nm')['order_rk'].nunique()

valid_regions = region_order_counts[region_order_counts >= 2].index

df['order_present'] = df['lvn_state_nm'].apply(lambda x: 1 if x in valid_regions else 0)

region_data = df.groupby('lvn_state_nm')['order_present'].value_counts().unstack(fill_value=0)

chi2_stat, p_value_3, dof, expected = stats.chi2_contingency(region_data)


print(f" p-value: {p_value_3}")

if p_value_3 < 0.05:
    print("Отклоняем нулевую гипотезу: Местоположение клиента (регион) влияет на частоту или вероятность покупки.")
else:
    print("Не отклоняем нулевую гипотезу: Местоположение клиента (регион) не влияет на частоту или вероятность покупки.")
```
